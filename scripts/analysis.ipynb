```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import des données
df = pd.read_csv("../Data/euromillion.csv", sep=";")

# Séparation des numéros et des étoiles
df[['Num1', 'Num2', 'Num3', 'Num4', 'Num5', 'Etoile1', 'Etoile2']] = df['Tirage'].str.split(' ', expand=True)
df[['Num1', 'Num2', 'Num3', 'Num4', 'Num5', 'Etoile1', 'Etoile2']] = df[['Num1', 'Num2', 'Num3', 'Num4', 'Num5', 'Etoile1', 'Etoile2']].astype(int)

# Calcul des fréquences des numéros et des étoiles
num_frequencies = pd.concat([df['Num1'], df['Num2'], df['Num3'], df['Num4'], df['Num5']]).value_counts()
etoile_frequencies = pd.concat([df['Etoile1'], df['Etoile2']]).value_counts()

# Visualisation
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Histogramme des fréquences des numéros
ax[0].bar(num_frequencies.index, num_frequencies.values)
ax[0].set_title('Fréquence des Numéros')
ax[0].set_xlabel('Numéro')
ax[0].set_ylabel('Fréquence')

# Histogramme des fréquences des étoiles
ax[1].bar(etoile_frequencies.index, etoile_frequencies.values)
ax[1].set_title('Fréquence des Etoiles')
ax[1].set_xlabel('Etoile')
ax[1].set_ylabel('Fréquence')

plt.tight_layout()
plt.show()
```

```python
# Affichage des 5 numéros les plus fréquents
num = [i for i in num_frequencies[:5].index]
etoile = [i for i in etoile_frequencies[:2].index]
print("Tirage le plus probable: ")
print(num+etoile)
```
